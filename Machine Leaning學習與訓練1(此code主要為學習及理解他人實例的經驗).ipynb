```python
import pandas as pd
#方差選擇
from sklearn.feature_selection import VarianceThreshold
#皮爾遜相關係數
from scipy.stats import pearsonr
#標準化
from sklearn.preprocessing import StandardScaler
#數據劃分
from sklearn.model_selection import train_test_split
#决策樹
from sklearn.tree import DecisionTreeClassifier,export_graphviz
#KNN
from sklearn.neighbors import KNeighborsClassifier
#網格搜索和交叉驗證
from sklearn.model_selection import GridSearchCV
#隨機森林
from sklearn.ensemble import RandomForestClassifier
```

```python
RawData = pd.read_csv('C:/Users/Tsim/Desktop/Machine Learning/winequality-white.csv',sep=';')
print(type(RawData))
print(RawData[:6])
```

```python
transfer = VarianceThreshold()
new_data = transfer.fit_transform(RawData)
print('new_data:\n', new_data, new_data.shape)
```

```python
data_title = RawData.columns.values.tolist()
r = []
for i in range(len(data_title)):
    for j in range(i+1, len(data_title)):
        print('%s 和 %s之間的皮爾遜相關係數為%f.'%(data_title[i], data_title[j],\
        pearsonr(RawData[data_title[i]], RawData[data_title[j]])[0]))
        r.append(pearsonr(RawData[data_title[i]], RawData[data_title[j]])[0])
```

```python
print([x for x in r if x >0.6])
```

```python
def Quality(e):
    if e>0 and e<=4:
        return 'inferior'
    elif e>4 and e<=7:
        return 'ordinary'
    else:
        return 'superior'

RawData['Quality'] = RawData['quality'].map(Quality)
RawData.drop(['quality'],axis=1,inplace=True)
RawData[:6]
```

```python
data2 = RawData.iloc[:,0:-1].values
data3 = RawData['Quality']

x_train,x_test,y_train,y_test = train_test_split(data2,data3,test_size =0.2,random_state = 6)

print(y_test)
```

```python
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)

estimator1 = KNeighborsClassifier()

param_dict = {'n_neighbors':[i for i in range(1,16)]}
estimator1 = GridSearchCV(estimator1,param_grid=param_dict,cv = 10)
estimator1.fit(x_train,y_train)

y_predict1 = estimator1.predict(x_test)
print('布爾對比:\n',y_predict1==y_test)

score1 = estimator1.score(x_test,y_test)
print('模型準確率:\n',score1)

print('最佳结果:\n',estimator1.best_estimator_)
```

```python
estimator = DecisionTreeClassifier(criterion='entropy')
estimator.fit(x_train,y_train)

y_predict = estimator.predict(x_test)

print('布爾對比:\n',y_predict==y_test)

score = estimator.score(x_test,y_test)
print('模型準確率:\n',score)
```
